```python
taxi_data = [feb_2020, mar_2020, apr_2020, may_2020, jun_2020] 

#Calculate proportion of missing data
total_data = 0
missing_data = 0
for df in taxi_data:
    missing_prop = (len(df[df.isna().any(axis=1)])/len(df))*100
    total_data = len(df) + total_data
    missing_data =  len(df[df.isna().any(axis=1)]) + missing_data
    print("total {} = {}".format(get_df_name(df), len(df)))
    print("total data = {}, missing_data = {}".format(total_data, missing_data))
    print("{} : {}% of missing data".format(get_df_name(df), missing_prop))
    if get_df_name(df) == get_df_name(taxi_data[-1]):
        print("percentage of missing data over all data = {}%".format(round((missing_data/total_data)*100, 3)))
```

### Analysis of Outlier

1. {tip_amount} - It will be unfair to remove the data simply because it's an outlier, so we are going to assume the 20% tipping culture of USA. which means we need to feature engineer a $$ percentagetip = (tip / fare) * 100 $$


```python
# Feature Engineering


def add_feature(df):
    df["tip_perc"] = (df["tip_amount"]/df["fare_amount"]) * 100

    # Calculates duration of ride in seconds
    df["duration"] = (df["tpep_dropoff_datetime"] - df["tpep_pickup_datetime"]).astype('timedelta64[s]')
    # in hours
    df["duration"] = df["duration"]/(3600)

    # average speed of trip (miles/h)
    df["avg_speed"] =  df["trip_distance"] / df["duration"]
    
    return df

    
```

```python
def clean_data(df, month):
    
    # initial assumptions of data
    
    #dates should be corresponding to 2020 - maybe don't remove convert year to correct year
    df = df[pd.DatetimeIndex(df["tpep_dropoff_datetime"]).year == 2020]
    df = df[pd.DatetimeIndex(df["tpep_pickup_datetime"]).year == 2020]
    
    # This makes sure the dates are corresponding to the correct month
    # according to user input "month" as a number with no padding eg "5" for may
    df = df[pd.DatetimeIndex(df["tpep_dropoff_datetime"]).month == month]
    df = df[pd.DatetimeIndex(df["tpep_pickup_datetime"]).month == month]
    
    # drop off time must not be before pick up time
    df = df[df["tpep_pickup_datetime"] < df["tpep_dropoff_datetime"]]
    
    #these values should always be postive
    df = df[df.trip_distance > 0]
    df = df[df.fare_amount > 0]
    df = df[df.tip_amount >= 0]
    
    # we assume that tolls_amount shouldn't be more than the fare_amount
    df = df[df.fare_amount > df.tolls_amount]
    
    # only taking payment type 1, 2, and 5
    df = df[(df.payment_type == 1.0 )| (df.payment_type == 2.0) | (df.payment_type == 5.0)]
    
    return df


def clean_data2(df):
    # we assume that the maximum speed a trip can take should be no more than the legal limit of 55mph on the high end
    df = df[df.avg_speed < 55]
    
    # tip percentage should follow the advised 20% however we would include it as 25%
    df = df[df.tip_perc < 25]
    
    # Duration of trip must not be > 10h as it is prohibited by the law
    #hence any trip lasting more than 10h is assumed as incorrect data
    df = df[df.duration < 10]
    
    return df

```

```python
# uses the clean function to do most of the cleaning, however it might not be fully cleaned
feb_2020 = clean_data(feb_2020, 2)
mar_2020 = clean_data(mar_2020, 3)
apr_2020 = clean_data(apr_2020, 4)
may_2020 = clean_data(may_2020, 5)
jun_2020 = clean_data(jun_2020, 6)

    


# for df in taxi_data:
#     df.boxplot(column = ["trip_distance"])
#     df.boxplot(column = ["fare_amount"])
#     df.boxplot(column = ["extra"])
# apr_2020.boxplot(column = ["tip_amount"])
#     df.boxplot(column = ["tolls_amount"])
#     df.boxplot(column = ["improvement_surcharge"])
#     df.boxplot(column = ["total_amount"])
#     df.boxplot(column = ["congestion_surcharge"])
    
```

### Identifying the outliers

As we can see there are quite a number of outliers according to the box plot. But the question is how many?


```python
def calc_quantiles(df_column):
    Q1 = df_column.quantile(0.25)
    Q3 = df_column.quantile(0.75)
    IQR = Q3 - Q1
    
    return Q1, Q3, IQR
```

```python
feb_q1, feb_q3, feb_iqr = calc_quantiles(feb_2020.tip_amount)
mar_q1, mar_q3, mar_iqr = calc_quantiles(mar_2020.tip_amount)
```

```python
# Find the number of outliers

#Finds the upper outliers
print("Number of upper outliers: ", len(feb_2020[feb_2020['tip_amount'] > feb_q3 + (1.5 * feb_iqr)]))

# Finds the lower outliers
print("Number of lower outliers: ", len(feb_2020[feb_2020['tip_amount'] < feb_q1 - (1.5 * feb_iqr)]))
```
